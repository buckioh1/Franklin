## Use of .apply():

### 1. Create a function that returns the mean, sum, mode, median, and range (separately)

```python
def math(vec):
    if vec.dtype != 'object':
        mean = np.nanmean(vec)
        mean_value = f"${mean:,.2f}"
        add = np.nansum(vec)
        add_value = f"${add:,.2f}"
        vals, counts = np.unique(vec, return_counts = True)
        index = np.argmax(counts)
        mode = vals[index]
        mode_value = f"${mode:,.2f}"
        median = np.nanmedian(vec)
        median_value = f"${median:,.2f}"
        range_num = np.nanmax(vec) - np.nanmin(vec)
        range_num_value = f"${range_num:,.2f}"
        return mean_value, add_value, mode_value, median_value, range_num_value
```

### 2. Run the function into your chosen data set using the .apply() method.

```python
df.apply(math).rename(index = {0: "Mean", 1: "Addition", 2: "Mode", 3: "Median", 4: "Range"})
```
