Looking at the variables and data types, let's use `Total Amount Reimbursed` as our dimention variable and `State` as our `groupby()` attribute.

````python
def summarize_data(g):
  return(
    pd.Series({
      "avg_Total_Reimbursed": g['Total Amount Reimbursed'].mean(),
      "min_Total_Reimbursed": g['Total Amount Reimbursed'].min(),
      "max_Total_Reimbursed": g['Total Amount Reimbursed'].max(),
      "sample_count": len(g),
      "distribution": g['Total Amount Reimbursed'].sum()/df['Total Amount Reimbursed'].sum(),
    })
  )

df_group = df.groupby("State").apply(summarize_data).sort_values('avg_Total_Reimbursed', ascending = False).head(10)
df_group = df_group.style.format({
    'avg_Total_Reimbursed': '${:,.2f}',
    'min_Total_Reimbursed': '${:,.2f}',
    'max_Total_Reimbursed': '${:,.2f}',
    'sample_count': '{:,.0f}',
    'distribution': '{:,.3%}',
})
df_group
````
